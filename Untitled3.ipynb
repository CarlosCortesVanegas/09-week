```python
import pandas as pd
fechas = pd.date_range(start='2023-01-01', periods=10, freq='D')
df = pd.DataFrame({'fechas': fechas})
df['año'] = df['fechas'].dt.year
df['mes'] = df['fechas'].dt.month
df['día'] = df['fechas'].dt.day
df['día_semana'] = df['fechas'].dt.day_name()
df['actividad'] = [
    'Reunión', 'Trabajo remoto', 'Desarrollo', 'Análisis', 'Reporte',
    'Planificación', 'Presentación', 'Documentación', 'Pruebas', 'Entrega'
]

```

# Usa un conjunto de datos con una columna de fechas e intenta:
### Calcular cuántos días han pasado desde una fecha determinada.
### Filtrar los datos por fechas.


```python
from datetime import datetime

# Creamos un DataFrame con fechas aleatorias
df_fechas = pd.DataFrame({
    'evento': ['Evento A', 'Evento B', 'Evento C'],
    'fecha': pd.to_datetime(['2022-12-01', '2023-03-15', '2024-01-20'])
})

# Calcular cuántos días han pasado desde una fecha
hoy = pd.Timestamp(datetime.today())
df_fechas['días_desde_evento'] = (hoy - df_fechas['fecha']).dt.days
df_fechas

# Filtrar datos por fechas (después del 01/01/2023)
df_filtrado = df_fechas[df_fechas['fecha'] > '2023-01-01']
df_filtrado
```

# Jerarquía de Fechas
# Descripción
### La jerarquía de fechas permite agrupar datos por niveles temporales como año, trimestre, mes, semana o día. Esto es clave para realizar análisis comparativos y visualizaciones en dashboards.

# Aplicabilidad
# Muy útil para:

# Dashboards ejecutivos

### Reportes periódicos (mensuales, trimestrales)

### Visualización de tendencias

### Modelos de pronóstico
# casos practicos

```python
import numpy as np

# Simulación de un DataFrame con ventas diarias
df = pd.DataFrame({
    'fecha': pd.date_range(start='2023-01-01', periods=90, freq='D'),
    'ventas': np.random.randint(100, 1000, size=90)
})

# Agrupación por mes
df['mes'] = df['fecha'].dt.to_period('M')
ventas_mensuales = df.groupby('mes')['ventas'].sum()
ventas_mensuales
```

# actividad 2
### Agrupa un dataset por trimestre y calcula estadísticas descriptivas.
### Visualiza las ventas promedio por mes.


```python
# Agrupar por trimestre y calcular estadísticas
df['trimestre'] = df['fecha'].dt.to_period('Q')
estadísticas_trimestrales = df.groupby('trimestre')['ventas'].describe()
estadísticas_trimestrales

# Visualizar ventas promedio por mes
import matplotlib.pyplot as plt

ventas_mensuales_promedio = df.groupby('mes')['ventas'].mean()
ventas_mensuales_promedio.plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title('Ventas Promedio por Mes')
plt.ylabel('Promedio de Ventas')
plt.xlabel('Mes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
